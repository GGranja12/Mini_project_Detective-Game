```python
import time

#suspects and murderer
suspects = ["Guilherme", "Charlotte", "Ruben", "Oeendrila"]
murderer = "Ruben"

#Suspect 
suspect_files = {
    1: "Guilherme, 25 years old: Assistant to Dr. Wiener. Although he was absent during the murder, he contributed a key idea that Dr. Wiener used to become wealthy. Guilherme did not receive recognition or compensation, which could have generated strong resentment.",
    2: "Ruben, 26 years old: Close friend and partner in secret projects with Dr. Wiener. Discrepancies in these projects could have motivated him to commit the crime.",
    3: "Oeendrila, 37 years old: Recent girlfriend of Dr. Wiener. She had a heated argument about the inheritance, as she would receive nothing, which might have impulsively driven her to act.",
    4: "Charlotte, 27 years old: Daughter and heiress of Dr. Wiener. She was facing financial difficulties, and her father refused to help her after she lost 2 million, giving her a strong motive to eliminate him and claim the inheritance."
}

#clues 
clues = {
    'living_room': {
        "clues": {
            1: "Kitchen knife: The murder weapon, found at the crime scene.",
            2: "Paint-stained gloves: Found in the living room, used by Guilherme.",
            3: "Expensive dinner receipt: Paid by Charlotte, suggesting her presence in the house.",
            4: "Financial note: Mentions an agreement with Dr. Wiener, appears to be from Guilherme.",
            5: "Broken wine glass: Found in the living room, suggests an attempt to clean the scene."
        },
        #suspicious Ruben, Charlotte, Guilherme, Oeendrila
    },
    'kitchen': {
        "clues": {
            1: "Charlotte's comb: Found in the kitchen, suggesting she was near the knife.",
            2: "Food remnants: Match Charlotte’s preferences.",
            3: "Toolbox: Used by Ruben, but could also have been used by Guilherme.",
            4: "Work gloves: Stained with paint, used by Guilherme.",
            5: "Spare keys: Ruben had access to restricted areas, but Charlotte also had copies."
        },
        #suspicious Charlotte, Guilherme
    },
    'bedroom': {
        "clues": {
            1: "Ruben's watch: Found near the bed, left in a hurry.",
            2: "Glasses on the floor: Worn by Oeendrila, possibly dropped during a struggle.",
            3: "Open safe: Documents of Ruben found inside, but there were also papers of Oeendrila.",
            4: "Financial note: Mentions an agreement with Dr. Wiener, appears to be from Guilherme.",
            5: "Mud trail: On the carpet near the bed, suggests someone with muddy shoes was recently present."
        },
        #suspicious Charlotte, Oeendrila, Guilherme
    },
    'garage': {
        "clues": {
            1: "Used tools: Match Ruben’s, but Guilherme’s tools are also present.",
            2: "Ruben's car: Signs of having been at the scene, but there are also traces of another vehicle.",
            3: "Paint-stained gloves: Ruben used them, but similar gloves were found in Oeendrila’s room.",
            4: "Stains on the floor: Chemical products associated with Ruben, but the composition matches Guilherme’s.",
            5: "Ruben's note: About a disagreement with Dr. Wiener, but notes also appear to be from Oeendrila."
        },
        #suspicious  Ruben, Guilherme, Oeendrila
    },
    'bathroom': {
        "clues": {
            1: "Stained towel: Could indicate an attempt to clean, but the stain pattern suggests it might have been used by Charlotte.",
            2: "Wine glass: Found in the bathroom, used to clean blood, suggests Oeendrila, known for her love of wine, might have been involved.",
            3: "Wet wipes: Possibly used by Ruben, but also found in Charlotte's bag.",
            4: "Chemical bottle: Used by Ruben in his projects, but similar bottles were also found in Oeendrila’s room.",
            5: "Cleaning note: Found in the bathroom, mentions cleaning tasks, but it is unclear to whom it belongs."
        },
       #suspicious Charlotte, Oeendrila
    }
}

#oom connections
room_connections = {
    "expedient_room": ["living_room"],
    "living_room": ["kitchen", "bedroom", "Accusatory room", "expedient_room"],
    "kitchen": ["living_room", "garage"],
    "garage": ["kitchen"],
    "bedroom": ["bathroom", "living_room"],
    "bathroom": ["bedroom"]
}

game_active = True
current_room = "living_room"  #starting room

#introduction game
def introduction(): 
    time.sleep(2)
    
    print("Welcome to Ludo inspired by Cluedo!")
    print(" ")
    time.sleep(2)
    print("In the heart of bustling Stockholm, 1964, stands the grand mansion of Dr. Norbert Wiener, a renowned scientist known for his groundbreaking research in artificial intelligence.")
    print(" ")
    time.sleep(2)
    print("Dr. Wiener was a figure of intrigue and controversy. His enigmatic projects in AI and data sparked curiosity and envy, creating a web of potential motives among those around him.")
    print(" ")
    time.sleep(2)
    print("One fateful night, a chilling cry shattered the silence of the mansion. Dr. Wiener was found dead in the living room, a shocking end to his life of accomplishments.")
    print(" ")
    time.sleep(2)
    print("You are part of the Swedish National Police, called in to solve this complex mystery. Your task is to investigate the mansion, gather clues, and piece together the puzzle to identify the murderer.")
    print(" ")
    time.sleep(3)
    print("Your Mission:")
    print("As the lead investigator, you must carefully explore each room, examine the clues, and question the suspect. When you feel ready, make your final accusation and reveal the murderer.")
    print(" ")

#suspects
def introduction2(): 
    time.sleep(2)
    print(" ")
    print("We arrive in front of the house and all the suspects are there, waiting for us.")
    time.sleep(2)
    print(" ")
    print("Here's what we know about them so far...")
    time.sleep(2)
    print("Guilherme, 25 years old: Assistant to Dr. Wiener. Although he was absent during the murder, he contributed a key idea that Dr. Wiener used to become wealthy. Guilherme did not receive recognition or compensation, which could have generated strong resentment.")
    time.sleep(2)
    print("Ruben, 26 years old: Close friend and partner in secret projects with Dr. Wiener. Discrepancies in these projects could have motivated him to commit the crime.")
    time.sleep(2)
    print("Oeendrila, 37 years old: Recent girlfriend of Dr. Wiener. She had a heated argument about the inheritance, as she would receive nothing, which might have impulsively driven her to act.")
    time.sleep(2)
    print("Charlotte, 27 years old: Daughter and heiress of Dr. Wiener. She was facing financial difficulties, and her father refused to help her after she lost 2 million, giving her a strong motive to eliminate him and claim the inheritance.")
    print(" ")

#expedient room
def review_files():
    global current_room  
    print("Reviewing suspect files.")
    while True:
        print("Enter a number (1-4) to view a file or 5 to go back:")
        option = input()
        if option.isdigit():
            option = int(option)
            if 1 <= option <= 4:
                print(suspect_files[option])
            elif option == 5:
                current_room = "living_room"  
                print("Going back to the living room...")
                return
            else:
                print("Invalid option. Try again.")
        else:
            print("Invalid input. Please enter a number.")

#nvestigat cclues
def explore_room():
    global current_room
    clues_in_rooms = clues   
    room_clues = clues_in_rooms[current_room]["clues"]
    print(f"You are in the {current_room.replace('_', ' ').title()}. What would you like to do?")
    print("1: Investigate clues, 2: Move to another room.")
    option = input()
    
    if option == "1":
        print("Choose a clue (1-5):")
        clue_option = input()
        if clue_option.isdigit():
            clue_option = int(clue_option)
            if 1 <= clue_option <= 5:
                print(room_clues[clue_option])
            else:
                print("Invalid option. Try again.")
                explore_room()
        else:
            print("Invalid input. Please enter a number.")
            explore_room()
    elif option == "2":
        move_to_room()
    else:
        print("Invalid option. Try again.")
        explore_room()

#move rooms
def move_to_room():
    global current_room
    available_rooms = room_connections[current_room]
    print("Available rooms to move to:")
    for i, room in enumerate(available_rooms, 1):
        print(f"{i}. {room.replace('_', ' ').title()}")
    
    print("Enter the number of the room you'd like to move to:")
    room_choice = input()
    
    if room_choice.isdigit():
        room_choice = int(room_choice)
        if 1 <= room_choice <= len(available_rooms):
            target_room = available_rooms[room_choice - 1]
            if target_room == "Accusatory room":
                make_accusation()
            else:
                current_room = target_room
                print(f"Moving to {target_room.replace('_', ' ').title()}...")
                game_loop()
        else:
            print("Invalid selection. Try again.")
            move_to_room()
    else:
        print("Invalid input. Please enter a number.")
        move_to_room()

#makeaccusation
def make_accusation():
    print("Who do you think is the murderer?")
    print(f"Choose from the following suspects: {suspects}")
    suspect_choice = input()

    if suspect_choice.capitalize() in suspects:
        if suspect_choice.capitalize() == murderer:
            final_narrative()  # Call final narrative
            global game_active
            game_active = False
        else:
            print(f"Sorry, your accusation is incorrect. The murderer is not {suspect_choice.capitalize()}.")
    else:
        print("Invalid suspect name.")

#Finalnarrative
def final_narrative():
    time.sleep(2)
    print(" ")
    print("Congratulations! You have correctly accused the murderer!")
    time.sleep(2)
    
    print("Ruben, furious about the discrepancies in the secret projects with Dr. Wiener, decided to attack him during a heated argument in the living room. He used a kitchen knife to commit the murder.")
    time.sleep(2)
    
    print("To divert attention, Ruben left Charlotte's comb in the kitchen, suggesting that she had been near the knife. In the bathroom, he placed a wine glass used by Oeendrila, filled with wine and with lipstick stains, trying to implicate Oeendrila, known for her love of wine.")
    time.sleep(2)
    
    print("Furthermore, Ruben spread paint and left paint-stained gloves in the living room to implicate Guilherme. However, despite the multiple false leads, Ruben appeared to be the least suspicious initially, as the other suspects also had very evident clues. This initial confusion led the authorities to doubt.")
    time.sleep(2)
    
    print("Finally, overwhelmed by the manipulated evidence and the weight of his guilt, Ruben confessed in a moment of desperation. He admitted that the murder of Dr. Wiener was an accident and that he had not intended to kill him.")
    time.sleep(2)

#ggameloop
def game_loop():
    global current_room
    while game_active:
        if current_room == "expedient_room":
            review_files()
        else:
            explore_room()

#start of the game
introduction()
introduction2()

while True:
    user_input = input("Are you ready to investigate? (yes/no): ")
    if user_input.lower() in ["yes", "y"]:
        print("Let's start!")
        break
    elif user_input.lower() in ["no", "n"]:
        print("Exiting the game...")
        exit()
    else:
        print("Invalid input. Please enter yes/no.")

#start the game loop
game_loop()

#End game
print("Thank you for playing the Ludo!")
```
